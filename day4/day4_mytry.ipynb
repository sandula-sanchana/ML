```python
import numpy as np
import seaborn as sns
import pandas as pd


df=pd.read_csv('adv.csv')

#print(df.head(5))

sns.set_style('whitegrid')

tv_data=df['TV'].values.reshape(-1,1)

sales_data=df['Sales'].values

#print(tv_data)
#print(sales_data)


np.random.seed(42)

indices=np.random.permutation(len(tv_data))

#print(indices)

x_test=tv_data[indices[:40]]
x_train=tv_data[indices[40:]]

y_test=sales_data[indices[:40]]
y_train=sales_data[indices[40:]]

ones=np.ones((len(x_train),1))

x_and_ones=np.concatenate((ones,x_train),axis=1)

xt=x_and_ones.T

model_weight=np.linalg.inv(xt@x_and_ones)@xt@y_train

print("Model Weights: ",model_weight)


```

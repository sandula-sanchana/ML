```python
np.random.seed(42)

indices=np.random.permutation(160)
indices2=np.random.permutation(160)


x_train = indices.reshape(-1, 1)
y_train = indices2.reshape(-1, 1)


ones=np.ones((160,1))

x_and_ones=np.concatenate((ones,x_train),axis=1)

x_and_ones_T=x_and_ones.T

first_part=x_and_ones_T@x_and_ones

first_part_inv=np.linalg.inv(first_part)


second_part=x_and_ones_T@y_train

model_weight=first_part_inv@second_part

print("Model Weights:\n",model_weight)
```

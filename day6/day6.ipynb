```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer

df=pd.read_csv('womens_clothing_ecommerce_reviews.csv')
print(df.head())

X=df['Review Text']
Y=df['sentiment']

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42, stratify=Y)



```

```python
vectorizer=CountVectorizer(stop_words='english')
X_train_bow=vectorizer.fit(X_train)
X_train_bow=vectorizer.transform(X_train) #return spares matrix

X_test_bow=vectorizer.transform(X_test)

model=LogisticRegression(max_iter=5000,solver="saga",random_state=42)

model.fit(X_train_bow,y_train)

print(model.coef_)

y_train_pred=model.predict(X_train_bow)
y_test_pred=model.predict(X_test_bow)

train_accu=(y_train,y_train_pred)
test_accu=(y_test,y_test_pred)

print(train_accu)
print(test_accu)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(stop_words='english')

X_train_bow = vectorizer.fit_transform(X_train)   # sparse matrix
X_test_bow  = vectorizer.transform(X_test)

model = LogisticRegression(max_iter=5000, solver="saga", random_state=42)
model.fit(X_train_bow, y_train)

y_train_pred = model.predict(X_train_bow)
y_test_pred  = model.predict(X_test_bow)

train_accu = accuracy_score(y_train, y_train_pred)
test_accu  = accuracy_score(y_test, y_test_pred)

print("Train accuracy:", train_accu)
print("Test accuracy:", test_accu)
```

```python
new_review = ["The fabric feels cheap and the fit was completely off. Very disappointed with this purchase."]

new_review_bow = vectorizer.transform(new_review)

prediction = model.predict(new_review_bow)

if prediction[0] == 1:
    print("Positive Review ðŸ˜Š")
else:
    print("Negative Review ðŸ˜ž")
```

```python
confusion_matrix=confusion_matrix(y_test,y_test_pred)
classification_report=classification_report(y_test,y_test_pred)

print(classification_report)
print(confusion_matrix)

#Summary-wise this model is good for positive reviews but not so good for negative reviews
```

```python
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression as lr

from sklearn.metrics import root_mean_squared_error as rmse

from sklearn.linear_model import Ridge


df=pd.read_csv("ridge_vs_ols.csv")

X = df.loc[:, "x1":"x20"] # also can be done bu drop column y
Y=df["y"]



x_train,x_test,y_train,y_test=train_test_split(
      X,Y,
      test_size=20,
        random_state=42
)


model=lr()
model.fit(x_train,y_train)

w0=model.intercept_
coefs=model.coef_

print("w0:",w0)
print("coefs:",coefs)


y_pred=model.predict(x_train)
y_test_pred=model.predict(x_test)


train_rmse=rmse(y_train,y_pred)
test_rmse=rmse(y_test,y_test_pred)

print("train rmse:",train_rmse)
print("test rmse:",test_rmse)




```

```python
Ridge_model=Ridge(alpha=0.56)#ALPHA IS THE REGULARIZATION PARAMETER

Ridge_model.fit(x_train,y_train)

w0_ridge=Ridge_model.intercept_
coefs_ridge=Ridge_model.coef_
print("w0 ridge:",w0_ridge)
print("coefs ridge:",coefs_ridge)

y_pred_ridge=Ridge_model.predict(x_train)
y_test_pred_ridge=Ridge_model.predict(x_test)

train_rmse_ridge=rmse(y_train,y_pred_ridge)
test_rmse_ridge=rmse(y_test,y_test_pred_ridge)
print("train rmse ridge:",train_rmse_ridge)
print("test rmse ridge:",test_rmse_ridge)
```

```python
import numpy as np
import pandas

import matplotlib.pyplot as plt

img = np.random.randint(0, 100, (5, 5))
x = img + 100
x = np.clip(x, 0, 255)

print("Original:\n", img)
print("\nAfter +100:\n", x)


plt.subplot(1, 2, 1)
plt.axis('off')
plt.title("Original")
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.axis('off')
plt.title("After Scalar Addition")
plt.imshow(x, cmap='gray', vmin=0, vmax=255)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


batch_1 = np.random.randint(0, 255, (4, 5, 5))
batch_2 = np.random.randint(0, 255, (4, 5, 5))


flattened_b1 = batch_1.reshape(4, -1)  # shape (4,25)
flattened_b2 = batch_2.reshape(4, -1)  # shape (4,25)


similarities = np.sum(flattened_b1 * flattened_b2, axis=1)  # shape (4,)
print("Dot product similarities:", similarities)


plt.figure(figsize=(9, 10))

for i in range(4):

    plt.subplot(4, 3, i * 3 + 1)
    plt.imshow(batch_1[i], cmap='gray', vmin=0, vmax=255)
    plt.title(f"Batch 1 - Image {i+1}")
    plt.axis('off')


    plt.subplot(4, 3, i * 3 + 2)
    plt.imshow(batch_2[i], cmap='gray', vmin=0, vmax=255)
    plt.title(f"Batch 2 - Image {i+1}")
    plt.axis('off')

    plt.subplot(4, 3, i * 3 + 3)
    plt.axis('off')
    plt.text(0.5, 0.5, f"Dot: {similarities[i]}", ha='center', va='center', fontsize=12)
    plt.title("Similarity")

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


#(nn)(n1)=(n1) , matrix * vector


# Original image (5x5)
x = np.random.randint(0, 255, (5, 5))

# Convert image to column vector (25x1)
v = x.flatten().reshape(25, 1)

# Transformation matrix (25x25)
M = 2.5 * np.identity(25)

#print(M)

# MATRIX  VECTOR
y = M @ v    # (25x1)

# Reshape back to image (5x5)
y_img = y.reshape(5, 5)

# Plot
plt.subplot(1, 2, 1)
plt.axis('off')
plt.title("Original")
plt.imshow(x, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.axis('off')
plt.title("After matrix  vector")
plt.imshow(y_img, cmap='gray', vmin=0, vmax=255)

plt.show()
```

```python
def sigmoid_threshold(x):
    s = 1 / (1 + np.exp(-x))
    return np.where(s < 0.65, 'False', 'True')

arr = np.array([0, 10, -1, 45, 99])
print(sigmoid_threshold(arr))
```

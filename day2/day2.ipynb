```python
import numpy as np
import matplotlib.pyplot as plt

img = np.random.randint(0, 100, (5, 5))
x = img + 100
x = np.clip(x, 0, 255)

print("Original:\n", img)
print("\nAfter +100:\n", x)


plt.subplot(1, 2, 1)
plt.axis('off')
plt.title("Original")
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.axis('off')
plt.title("After Scalar Addition")
plt.imshow(x, cmap='gray', vmin=0, vmax=255)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

img = np.random.randint(0, 100, (5, 5))
x = img+ np.array([0 ,50,100,150,200])
x = np.clip(x, 0, 255)

print("Original:\n", img)
print("\nAfter broadcasting:\n", x)


plt.subplot(1, 2, 1)
plt.axis('off')
plt.title("Original")
plt.imshow(img, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 2, 2)
plt.axis('off')
plt.title("After broadcasting")
plt.imshow(x, cmap='gray', vmin=0, vmax=255)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

imgA = np.random.randint(0, 100, (5, 5))
imgB=np.random.randint(0, 100, (5, 5))
x = imgA + imgB
x = np.clip(x, 0, 255)

print("A:\n", imgA)
print("\nB:\n", imgB)
print("\nAfter A+B:\n", x)


plt.subplot(1, 3, 1)
plt.axis('off')
plt.title("A")
plt.imshow(imgA, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 3, 2)
plt.axis('off')
plt.title("B")
plt.imshow(imgB, cmap='gray', vmin=0, vmax=255)

plt.subplot(1, 3, 3)
plt.axis('off')
plt.title("matrix Addition A+B")
plt.imshow(x, cmap='gray', vmin=0, vmax=255)

plt.show()
```

```python
import numpy as np

img_A = np.random.randint(0, 256, (10,10,3))
img_B = np.random.randint(0, 256, (10,10,3))
img_C = np.random.randint(0, 256, (10,10,3))


dotP_A_B = np.dot(img_A.flatten(), img_B.flatten())
dotP_A_C = np.dot(img_A.flatten(), img_C.flatten())

if dotP_A_B < dotP_A_C:
    print("A is more similar to C")
else:
    print("A is more similar to B")
```

```python
import numpy as np
movie_A=[2.5,0.5]
movie_B=[10,5]

user=[5,1]

user_andA=(np.dot(user,movie_A)/(np.linalg.norm(user)*np.linalg.norm(movie_A)))
user_andB=(np.dot(user,movie_B)/(np.linalg.norm(user)*np.linalg.norm(movie_B)))

print(f'user_andA: {user_andA}')
print(f'user_andB: {user_andB}')

if user_andA > user_andB:
    print("A is best match for user")
else:
    print("B is best match for user")
```
